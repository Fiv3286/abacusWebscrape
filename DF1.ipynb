````python
# import plain HTML
pageone = {}
for filename in os.listdir(PAGE_FOLDER_PATH):
  with open(PAGE_FOLDER_PATH + filename) as openfile:
    pageone[filename.split(".")[0].split(" ")[0]] = openfile.read()

list(pageone.keys()) # list pre-loaded pages to be scaped for their offers
````

````python
from bs4 import BeautifulSoup
import re
from collections import Counter


# the "magic" happens here.
# written by trial & error. Divs were identified using F12 inspect tool.
def get_top_listings(onepager):
  soupified = BeautifulSoup(onepager, 'html.parser')
  listings = soupified.find_all("div", {"class": "src-listing p-1 pb-0 flex flex-col justify-between w-full rounded-md border-solid border-[1px] !border-border !bg-white hover:!bg-hover hover:!border-abacus2 !mt-0"})
  listings_souped = []
  for listing in listings:
    listing_dict = {}
    seller_info = listing.find("div", {"class" : "flex flex-row items-center justify-center 2xl:justify-start gap-1 px-1"})
    feedback = seller_info.find("div", {"class": "absolute hidden group-hover:block anim anim-fadeIn -mt-20 left-1/2 -translate-x-1/2 w-max bg-abacus px-2 py-1 rounded text-white max-w-lg text-center"})
    fbs = [x.get_text().strip() for x in feedback.find_all("b")] if feedback != None else [0, 0, 0] #Feed Back Summary if exists. Patch fix so stuff doesn't break

    name = seller_info.find("div", {"class" : "flex flex-wrap items-center gap-1"}).find("a").get_text()

    what = listing.find("a").get_text().strip()
    # clean up the string some more
    what = re.sub(' +', ' ', what)
    listing_dict["what"] = what
    listing_dict["seller"] = name
    listing_dict["seller_rating"] = fbs
    listings_souped.append(listing_dict)

  return listings_souped


thedata = {}
for page in list(pageone.keys()):
  top_listing = get_top_listings(pageone[page])
  thedata[page] = top_listing
  print(page + ", " + str(len(top_listing)) + " top listings")

  sellers = [x['seller'] for x in top_listing]
  sellerNumber = [(string, count) for string, count in Counter(sellers).items()]
  sellerSorded = sorted(sellerNumber, key=lambda x: x[1], reverse=True)
  for pair in sellerSorded:
    seller_offers = [listing for listing in top_listing if listing['seller'] == pair[0]]
    print(f"  seller: {pair[0]} with {pair[1]} offers in top 30. rated  {seller_offers[0]['seller_rating'][0]} over {seller_offers[0]['seller_rating'][2]} reviews, {seller_offers[0]['seller_rating'][1]}")
    # seller offers
    for offer in seller_offers:
      print("     " + "".join(offer['what'].split()))
  print("")


````
